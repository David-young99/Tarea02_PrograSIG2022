## Primera sección

#### Carga de todas las librerías necesarias para la elaboración del programa

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import plotly.express as px #?
import plotly.figure_factory as ff #?

import plotly.graph_objects as go

import numpy as np
```

#### Carga de los archivos csv y depuración de los archivos para trabajar con mayor comodidad

```python
covid_positivos_crudos = pd.read_csv("datos/covid/05_30_22_CSV_POSITIVOS.csv", sep=";", encoding="iso-8859-1") #carga de los datos positivos covid
covid_generales_crudos = pd.read_csv("datos/covid/05_30_22_CSV_GENERAL.csv", sep=";", encoding="iso-8859-1") #carga de los datos generales covid

#Depurando los archivos 

#Limitando la carga de columnas que vamos a utilizar en ambos archivos
covid_positivos_crudos = covid_positivos_crudos[["provincia", "canton", "30/05/2022"]]
covid_generales_crudos = covid_generales_crudos[["FECHA", "hom_posi", "muj_posi", "menor_posi", "adul_posi", "am_posi", "nue_posi"]]

# Eliminación valores nulos
covid_positivos_crudos = covid_positivos_crudos.dropna(how='all')
covid_generales_crudos = covid_generales_crudos.dropna(how='all')

# Eliminación de fila con canton=="Otros"
covid_positivos_crudos = covid_positivos_crudos[covid_positivos_crudos["canton"] != "Otros"]


#cambiando los nombres de las columnas
covid_positivos = covid_positivos_crudos.rename(columns={"30/05/2022": "positivos"})

covid_generales = covid_generales_crudos.rename(
    columns={
        "FECHA": "fecha",
        "hom_posi":"hom_casos",
        "muj_posi": "muj_casos",
        "menor_posi": "menores_casos",
        "adul_posi": "adultos_casos",
        "am_posi": "adultosM_casos",
        "nue_posi": "nuevos_casos"
    }
)
```

```python
covid_generales
```

#### Creación del gráfico

```python
# Generación de la tabla únicamente con los datos Covid general

headerColor = 'grey'
rowEvenColor = 'lightgrey'
rowOddColor = 'white'

table1 = go.Figure(data=[go.Table(header=dict(values=list(covid_generales.columns),
                            line_color='darkslategray',
                            fill_color=headerColor,
                            align=['left','center'],
                            font=dict(color='white', size=10),
                                    
                                 ),
                              cells=dict(values=[
                                covid_generales.fecha,
                                covid_generales.hom_casos, 
                                covid_generales.muj_casos, 
                                covid_generales.menores_casos,
                                covid_generales.adultos_casos,
                                covid_generales.adultosM_casos,
                                covid_generales.nuevos_casos
                              ],
                            line_color='darkslategray',
                            # Colores intercalados
                            fill_color = [[rowOddColor,rowEvenColor,rowOddColor]*817],
                            align = ['left', 'center'],
                            font = dict(color = 'darkslategray', size = 9),
                                        
                                    ))
                     ])

table1.update_layout(
    margin=dict(l=20, r=20, t=20, b=20)
)

table1.show()
```

## Segunda sección

#### Creación del gráfico de barras
Se crea con los casos positivos registrados en las fechas respectivas

```python
## Elaboración del gráfico de barras por fecha registrado

covid_generales_nuevos = covid_generales[["fecha", "nuevos_casos"]]

# Creación del gráfico

graph_bar = covid_generales_nuevos.plot.bar(x="fecha", y="nuevos_casos", color="red", figsize=(12, 10))

plt.title("Cantidad de casos positivos nuevos por día (fecha escalada)")
plt.xticks(np.arange(0, 817, step=45)) #utilicé la libería Num Py para acomodar las labels del eje x


plt.ylabel("Cantidad de casos")
plt.xlabel("Fechas", labelpad=9)
```

## Tercera sección

#### Creación del gráfico de líneas

Se crea con la cantidad de casos acumulados para hombres y otro para mujeres

```python
# Reducción de columnas
covid_general_hombres_mujeres_positivos = covid_generales[["fecha", "hom_casos", "muj_casos"]]
```

```python


# Gráfico de líneas
covid_general_hombres_mujeres_positivos.plot.line(x="fecha", 
                        y=["hom_casos", "muj_casos"],
                        color={"hom_casos": "blue", 
                               "muj_casos": "pink"},
                        figsize=(12, 10),
                        linewidth=5
                       )

plt.title("Evolución a través del tiempo de los casos positivos acumulados de hombres y de los casos positivos acumulados de mujeres.")
plt.xlabel("Fecha")
plt.ylabel("Casos")


```

## Cuarta sección

#### Creación del gráfico de líneas
Se realiza con la fecha, con los datos de los menores, adultos y adultos mayores

```python
# Reducción de columnas
covid_general_menores_adultos_adulmayores = covid_generales[["fecha", "menores_casos", "adultos_casos", "adultosM_casos"]]
```

```python


# Gráfico de líneas
covid_general_menores_adultos_adulmayores.plot.line(x="fecha", 
                        y=["menores_casos", "adultos_casos", "adultosM_casos"],
                        color={"menores_casos": "green", 
                               "adultos_casos": "black",
                               "adultosM_casos": "red"},
                        figsize=(12, 10),
                        linewidth=5
                       )

plt.title("Evolución a través del tiempo de los casos positivos acumulados de menores, los casos positivos acumulados de adultos y de los casos positivos acumulados de adultos mayores.")
plt.xlabel("Fecha")
plt.ylabel("Casos")

```
