#### Carga de todas las librerías necesarias para la elaboración del programa

```python
import pandas as pd
import seaborn as sns
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt

import plotly.express as px #?
import plotly.figure_factory as ff #?

import plotly.graph_objects as go
```

#### Carga de los archivos csv y depuración de los archivos para trabajar con mayor comodidad

```python
covid_positivos_crudos = pd.read_csv("datos/covid/05_30_22_CSV_POSITIVOS.csv", sep=";", encoding="iso-8859-1") #carga de los datos positivos covid
covid_generales_crudos = pd.read_csv("datos/covid/05_30_22_CSV_GENERAL.csv", sep=";", encoding="iso-8859-1") #carga de los datos generales covid

#Depurando los archivos 

#Limitando la carga de columnas que vamos a utilizar en ambos archivos
covid_positivos_crudos = covid_positivos_crudos[["cod_provin", "cod_canton", "provincia", "canton", "30/05/2022"]]
covid_generales_crudos = covid_generales_crudos[["FECHA", "hom_posi", "muj_posi", "menor_posi", "adul_posi", "am_posi", "nue_posi"]]

# Eliminación valores nulos
covid_positivos_crudos = covid_positivos_crudos.dropna(how='all')
covid_generales_crudos = covid_generales_crudos.dropna(how='all')

# Eliminación de fila con canton=="Otros"
covid_positivos_crudos = covid_positivos_crudos[covid_positivos_crudos["canton"] != "Otros"]

# Est

#cambiando los nombres de las columnas
covid_positivos_crudos = covid_positivos_crudos.rename(columns={"30/05/2022": "positivos"})

covid_generales = covid_generales_crudos.rename(
    columns={
        "FECHA": "fecha",
        "hom_posi":"hom_casos",
        "muj_posi": "muj_casos",
        "menor_posi": "menores_casos",
        "adul_posi": "adultos_casos",
        "am_posi": "adultosM_casos",
        "nue_posi": "nuevos_casos"
    }
)
```

#### Creación del gráfico

```python
# Generación de la tabla únicamente con los datos Covid general

headerColor = 'grey'
rowEvenColor = 'lightgrey'
rowOddColor = 'white'

table1 = go.Figure(data=[go.Table(header=dict(values=list(covid_generales.columns),
                            line_color='darkslategray',
                            fill_color=headerColor,
                            align=['left','center'],
                            font=dict(color='white', size=10),
                                    
                                 ),
                              cells=dict(values=[
                                covid_generales.fecha,
                                covid_generales.hom_casos, 
                                covid_generales.muj_casos, 
                                covid_generales.menores_casos,
                                covid_generales.adultos_casos,
                                covid_generales.adultosM_casos,
                                covid_generales.nuevos_casos
                              ],
                            line_color='darkslategray',
                            # Colores intercalados
                            fill_color = [[rowOddColor,rowEvenColor,rowOddColor]*817],
                            align = ['left', 'center'],
                            font = dict(color = 'darkslategray', size = 11)
                                        
                                    ))
                     ])

table1.show()
```
